```python
# Columns to convert to dummy variables
columns_to_convert = [
    'can work long time before system?', 'self-learning capability?',
    'Extra-courses did', 'talenttests taken?', 'olympiads',
    'Taken inputs from seniors or elders', 'interested in games',
    'In a Realtionship?', 'worked in teams ever?', 'Introvert', 'Job/Higher Studies?',
    'Gentle or Tuff behaviour?', 'Management or Technical', 'Salary/work', 'hard/smart worker'
]

# Convert columns to dummy variables
dummy_columns = pd.get_dummies(df[columns_to_convert])

# Concatenate the dummy columns with the original DataFrame
df = pd.concat([df, dummy_columns], axis=1)

# Drop the original columns (optional)
df.drop(columns=columns_to_convert, inplace=True)
```

```python
# Convert object type columns to category type
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = df[col].astype('category')
```

```python
from sklearn.preprocessing import LabelEncoder

# Find columns with more than two unique categories
multi_category_columns = [col for col in df.columns if df[col].dtype == 'category' and len(df[col].cat.categories) > 2]

# Apply Label Encoding to columns with multiple unique categories
label_encoder = LabelEncoder()
for col in multi_category_columns:
    df[col] = label_encoder.fit_transform(df[col])
```

```python
from imblearn.over_sampling import SMOTE

X = df[['Acedamic percentage in Operating Systems', 'percentage in Algorithms', 'Percentage in Programming Concepts', 'Percentage in Software Engineering', 'Percentage in Computer Networks', 'Percentage in Electronics Subjects', 'Percentage in Computer Architecture', 'Percentage in Mathematics', 'Percentage in Communication skills', 'Type of company want to settle in?', 'Interested Type of Books', 'coding skills rating', 'Interested subjects', 'Logical quotient rating', 'Hours working per day', 'certifications', 'workshops']]
y = df['Suggested Job Role']
```

```python
smote = SMOTE(random_state=42)
```

```python
# Resample the dataset using SMOTE
X_resampled, y_resampled = smote.fit_resample(X, y)
```

```python
# Display the new class distributions after applying SMOTE
print("Class distribution after SMOTE:")
print(pd.Series(y_resampled).value_counts())
```

# Decision Tree Classifier

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.metrics import accuracy_score
```

```python
# Splitting the dataset into training and testing sets (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)
```

```python
# Decision Tree Classifier
decision_tree = DecisionTreeClassifier(random_state=42)
decision_tree.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

prediction = decision_tree.predict(X_test)

print(classification_report(y_test, prediction))
print('\n')
print(confusion_matrix(y_test, prediction))
```

```python
# Predictions
y_pred_dt = decision_tree.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

# Calculate accuracy score for Decision Tree
accuracy_dt = accuracy_score(y_test, y_pred_dt)
print(f"Accuracy (Decision Tree): {accuracy_dt:.4f}")
```
